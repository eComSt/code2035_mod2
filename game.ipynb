```python
from tkinter import *
from random import randint

window = Tk()
window.geometry("600x600")

class Clay:
    image = PhotoImage(file=r'elements/free-icon-pottery-7942410.gif').subsample(4, 4)

class Water:
    image = PhotoImage(file=r"elements/free-icon-water-drop-4246703.gif").subsample(4,4)

class Fire:
    image = PhotoImage(file=r"elements/free-icon-fire-9509865.gif").subsample(4,4)
    def __add__(self,other):
        if isinstance(other,Earth):
            return Clay

class Wind:
    image = PhotoImage(file=r"elements/wind.gif").subsample(4,4)

class Earth:
    image = PhotoImage(file=r"elements/ground.gif").subsample(4,4)
    def __add__(self,other):
        if isinstance(other,Fire):
            return Clay

canvas = Canvas(window,width = 600,height = 600)
canvas.pack()

elements = [Fire(),Water(),Wind(),Earth()]
for elem in elements:
    img = canvas.create_image(randint(50,550),randint(50,550),image = elem.image)

def move(event):
    images_id = canvas.find_overlapping(event.x,event.y,event.x+10,event.y+10)
    if len(images_id)==2:
       elem_id_1 = images_id[0]
       elem_id_2 = images_id[1]
       element_1 = elements[elem_id_1-1]
       element_2 = elements[elem_id_2-1]
       new_element = element_1+element_2
       if new_element not in elements:
            canvas.create_image(randint(50,550),randint(50,550),image = new_element.image)
            elements.append(new_element)
    canvas.coords(images_id,event.x,event.y)

window.bind("<B1-Motion>",move)
window.mainloop()
```
